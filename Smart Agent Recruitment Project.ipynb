# Problem Statement:
## Smart Agent Recruitment Challenge

FinMan is a Financial Distribution company. Over the last 10 years, they have created an offline distribution channel across India. They sell Financial products to consumers by hiring agents in their network. These agents are freelancers and get commission when they make a product sale.


Overview of FinMan On-boarding process
The Managers at FinMan are primarily responsible for recruiting agents. Once a manager has identified a potential applicant, the would explain the business opportunity to the agent. Once the agent provides the consent, an application is made to FinMan to become an agent. This date is known as application_receipt_date.


In the next 3 months, this potential agent has to undergo a 7 day training at the FinMan branch (about Sales processes and various products) and clear a subsequent examination in order to become a FinMan agent.


### The problem - Who are the best agents?
As is obvious in the above process, there is a significant investment which FinMan makes in identifying, training and recruiting these agents. However, there are a set of agents who do not bring in the expected resultant business. 

Finman has approached you and asked you to  to analyse and observe the overall past recruitment data to help your client derive meaningful insights.


# What To Do:
As we saw in the problem statement that the client is a Financial Distribution Company which sells their products through a distribution channel across India.
```python
manager_age_days = ds['Application_Receipt_Date'] - ds['Manager_DoB']
manager_age_years = manager_age_days.dt.days//365

manager_exp_days = ds['Application_Receipt_Date'] - ds['Manager_DOJ']
manager_exp_years = manager_exp_days.dt.days//365
```

# Categorical-Categorical Relation
Once we have checked the numerical-numerical relationship of the features, we can now look at the categorical-categorical relationship and then draw insights from them.

We can generate various hypothesis and test to see if they are true or not.
* Are the applicants who were hired by a manager more likely to become a FinMAn agent?
* Are Male applicants more likely to be selected as FinMan Agents then Female Applicants?
* Do married applicants become FinMan agents more as compared to single or any other marital status?
* Do female managers gets more agents than male managers?
* Are graduate applicants more likely to become an FinMan agent than other applicants?
* Do managers at highest level are hiring more FinMAn agents?
* Are the probatioary managers getting less FinMan agents?
* Does age of the applicant have any significant difference in them becoming an agent?

**1. Are the applicants who were hired by a manager more likely to become a FinMAn agent?**

```python
def cat_cat_testing(data1, col_name1, col_name2):
    plt.figure(figsize = (6, 4), dpi = 120)
    sns.countplot(data = data1, x = col_name1, hue = col_name2)
    plt.title(f"{data1.groupby([col_name1])[col_name2].value_counts(normalize = True)}", fontsize = 9)
    plt.show()
```

```python
cat_cat_testing(ds, 'Hired_By_Managers', 'Business_Sourced')
```

Whether the applicants were hired by manager or not doesn't have any difference in them being sourced as FinMan agent.

**2. Are Male applicants more likely to be selected as FinMan Agents then Female Applicants?**

```python
cat_cat_testing(ds, 'Applicant_Gender', 'Business_Sourced')
```

Female applicants are more likely to be selected as the FinMan agent than male applicants.

**3. Do married applicants become FinMan agents more as compared to single or any other marital status?**

```python
cat_cat_testing(ds, 'Applicant_Marital_Status', 'Business_Sourced')
```

Marital Status of the applicants doesnot make any difference whether the person will be hired or not.

**4. Do female managers gets more agents than male managers?**

```python
cat_cat_testing(ds, 'Manager_Gender', 'Business_Sourced')
```

Female managers hire more applicants who were selected to become FinMan agents.

**5. Are graduate applicants more likely to become an FinMan agent than other applicants?**

```python
cat_cat_testing(ds, 'Applicant_Qualification', 'Business_Sourced')
```

**6. Do managers with current highest designation are hiring more FinMAn agents?**

**10. Does age of the applicant have any significant difference in them becoming an agent?**

```python
# categorising applicants age into groups
app_age = ds[['Applicant_Age','Business_Sourced']][:]
app_age['age_group'] = ''
app_age['age_group'][app_age['Applicant_Age'].isnull()] = np.NaN
app_age['age_group'][app_age['Applicant_Age']>=80] = 'very old'
app_age['age_group'][(app_age['Applicant_Age']<80) & (app_age['Applicant_Age']>=60)] = 'senior citizen'
app_age['age_group'][(app_age['Applicant_Age']<60) & (app_age['Applicant_Age']>=18)] = 'adult'
app_age['age_group'][app_age['Applicant_Age']<18] = 'young'
```

```python
cat_cat_testing(app_age, 'age_group', 'Business_Sourced')
```
